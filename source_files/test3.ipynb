```python
import PyPDF2
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

def extract_text_from_pdf(pdf_path):
    pdf_file = open(pdf_path, 'rb')
    pdf_reader = PyPDF2.PdfReader(pdf_file)
    num_pages = len(pdf_reader.pages)

    text = ""
    for page_num in range(num_pages):
        page = pdf_reader.pages[page_num]  # Acessa a página pelo índice
        text += page.extract_text()

    pdf_file.close()
    return text

def preprocess_text(text):
    tokens = word_tokenize(text.lower())
    stop_words = set(stopwords.words('portuguese'))
    cleaned_tokens = [token for token in tokens if token.isalnum() and token not in stop_words]
    cleaned_text = ' '.join(cleaned_tokens)  # Adicione espaço entre os tokens
    return cleaned_text

pdf_path = 'BUCO PMRN.pdf'
extracted_text = extract_text_from_pdf(pdf_path)
preprocessed_text = preprocess_text(extracted_text)
```

```python
import re
import json
texto_questoes = extracted_text
# Texto com as questões


# Dividir o texto em questões
questoes = re.split(r"Questão \d+", texto_questoes)[1:]

# Lista para armazenar as questões em formato JSON
questoes_json = []

# Iterar sobre as questões
for i, questao in enumerate(questoes, start=1):
    # Extrair número da questão
    num_questao = f"Questão {i}"
    # Extrair enunciado da questão
    enunciado = questao.split('\n', 1)[1].strip()
    # Extrair alternativas
    alternativas = re.findall(r"([A-E]\)) (.*?)$", questao, re.MULTILINE)
    # Criar dicionário para a questão atual
    questao_dict = {
        "Número": num_questao,
        "Enunciado": enunciado,
        "Alternativas": alternativas
    }
    # Adicionar questão ao JSON
    questoes_json.append(questao_dict)

# Salvar JSON em um arquivo
with open("questoes.json", "w", encoding="utf-8") as json_file:
    json.dump(questoes_json, json_file, indent=4, ensure_ascii=False)

print("JSON criado com sucesso!")
```
